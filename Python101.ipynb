## Functions

# White Spaces are important!

```python
def SquareIt(x):
    return x * x

print(SquareIt(2))
```

```python
#You can pass functions around as parameters
def DoSomething(f, x): 
    return f(x)

b= DoSomething(SquareIt, 3)
#print(DoSomething(SquareIt, 3))
b
```

```python
#Lambda functions let you inline simple functions
print(DoSomething(lambda x: x * x * x, 3))

DoSomething(lambda x: x * x * x, 3)
```

## Looping

```python
listOfItems = [1, 2, 3, 4, 5, 6,]


for items in listOfItems:
    print(items)
    if (items % 2 == 0):
        print("is even dude:)")
    else:
        print("is odd :")
        
print ("All done.")
        
```

```python
for x in range(10):
    print(x)
```

```python
for x in range(10):
    if (x == 1):
        continue
    if (x > 3):
        break
    print(x)
```

```python
x = 0
while (x <= 10):
    print(x)
    
    
```
